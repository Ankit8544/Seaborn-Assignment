Q.No-01    Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans :-

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generating random data
np.random.seed(42)
data = {
    'X': np.random.randn(100),
    'Y': np.random.randn(100) * 2,
    'Category': np.random.choice(['A', 'B', 'C'], 100),
    'Time': np.arange(1, 101)
}
df = pd.DataFrame(data)
df
```

Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

1. **Scatter Plot:**
   Use: A scatter plot is used to visualize the relationship between two numerical variables. It shows individual data points as dots, making it easy to identify patterns, clusters, or correlations between the variables.

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', data=df)
plt.title('Scatter Plot')
plt.show()
```

2. **Line Plot:**
   Use: A line plot displays data points connected by straight lines. It is suitable for visualizing trends and changes in data over time or any other ordered variable.

```python
# Line Plot
plt.figure(figsize=(8, 6))
sns.lineplot(x='Time', y='Y', data=df)
plt.title('Line Plot')
plt.show()
```

3. **Bar Plot:**
   Use: A bar plot is used to represent categorical data with rectangular bars whose lengths are proportional to the values they represent. It is useful for comparing data between different categories.

```python
# Bar Plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Y', data=df)
plt.title('Bar Plot')
plt.show()
```

4. **Histogram:**
   Use: A histogram is used to show the distribution of a single numerical variable. It breaks the data into bins and displays the frequency or count of data points in each bin.

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['X'], bins=10)
plt.title('Histogram')
plt.show()
```

5. **Heatmap:**
   Use: A heatmap is a 2D representation of data where the values are represented using colors. It is commonly used to visualize correlation matrices or to display data on a grid.

```python
# Heatmap
numerical_df = df.drop('Category', axis=1)
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.tight_layout()
plt.show()
```

Q.No-02     Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions. 

    Note:  timepoint, signal, event, and region are columns in the fmri dataset.

Ans :-

```python
Fmri = sns.load_dataset('fmri')
Fmri
```

```python
sns.lineplot(data=Fmri, x="timepoint", y="signal", hue="event", style="region")

plt.title("fMRI Signal for different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Events")
plt.show()
```

Q.No-06    Use the "flights" dataset from seaborn to plot a heatmap.

Ans :-

```python
Flights = sns.load_dataset('flights')
Flights
```

```python
Flight_pivot = Flights.pivot_table(index="month", columns="year", values="passengers")

sns.heatmap(Flight_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Passenger Flights from 1949 to 1960")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
